```python
from sklearn.preprocessing import StandardScaler # Better to work with bayes classifier

scaler = StandardScaler()

X_train_sc = scaler.fit_transform(X_train.reshape(X_train.shape[0], -1))
y_train = y_train.reshape(-1, )
X_test_sc = scaler.fit_transform(X_test.reshape(X_test.shape[0], -1))
y_test = y_test.reshape(-1, )
```

```python
from sklearn.neighbors import KernelDensity

def KDE_param(X_train, y_train):
    n_classes = len(np.unique(y_train))

    kde_models = [] # I will have 1 result for each class

    for class_label in range(n_classes):

        X_class = X_train[y_train == class_label] # Taking the X of the class.

        #kde = KernelDensity(bandwidth = 0.1, kernel = "gaussian")
        kde = KernelDensity(bandwidth = 1, kernel = "gaussian")
        #kde = KernelDensity(bandwidth = 10, kernel = "gaussian")
        #kde = KernelDensity(bandwidth = "silverman", kernel = "gaussian")
        
        kde.fit(X_class) # Creating a KDE model for each class
        
        kde_models.append(kde)

    return kde_models

#m1 = KDE_param(X_train_sc, y_train)
```

```python
%%time

def bayes_classifier_kde(X_test, kde_models):
    probs = np.zeros((X_test.shape[0], len(kde_models))) # Creating a matrix of logs for each sample.

    for i, kde in enumerate(kde_models):
        """
        For each X, i will define the probability of X being of a class
        
        """
        probs[:, i] = kde.score_samples(X_test)

    y_pred = np.argmax(probs, axis = 1) # Taking the class predicted

    return y_pred

#y_pred1 = bayes_classifier_kde(X_test_sc, m1)
```

### Applying in the main components :

```python
from scipy.stats import skew, kurtosis
from sklearn.decomposition import PCA
```

```python
def compute_general_components(images):
    """
    Compute the components to apply the TSNE
    """
    std_ = np.zeros(images.shape[0])
    mean_ = np.zeros(images.shape[0])
    skew_ = np.zeros(images.shape[0])
    kurto_ = np.zeros(images.shape[0])
    for count, img in enumerate(images):
        std_[count] = np.std(img)
        mean_[count] = np.mean(img)
        skew_[count] = skew(img.flatten())
        kurto_[count] = kurtosis(img.flatten())

    return np.concatenate((std_.reshape(-1, 1), mean_.reshape(-1, 1), skew_.reshape(-1, 1), kurto_.reshape(-1, 1)), axis = 1)
```

```python
main_components = compute_general_components(X_train_sc)
```
